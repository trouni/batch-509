```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset('tips')
# tips_df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
tips_df.shape
```

```python
# seaborn_distplot = plt.subplot(1, 2, 2)
# seaborn_distplot.set_title('Seaborn')
sns.distplot(tips_df['total_bill'])
```

```python
tips_df.head(5)
```

```python
sns.countplot(x='day', data=tips_df, hue='time')
```

```python
sns.catplot(x='day', y='total_bill', data=tips_df, kind="violin")
```

```python
g = sns.FacetGrid(tips_df, col="day", row="sex", hue="smoker") # create grid
g.map(sns.distplot, "total_bill") # plot in each grid element
```

```python
sns.set(style='white')
sns.pairplot(tips_df, height=2)
plt.show()
```
