### Data selection

```python
data_list = [
    [ 0,  1,  2,  3,  4],
    [10, 11, 12, 13, 14],
    [20, 21, 22, 23, 24],
    [30, 31, 32, 33, 34],
    [40, 41, 42, 43, 44],
]
data_np = np.array(data_list)
data_np
```

```python
data_list[2][1:4]
```

```python
my_list = [
    [6, 5],
    [1, 3],
    [5, 6],
    [1, 4],
    [3, 7],
    [5, 8],
    [3, 5],
    [8, 4],
]
```

```python
sums = []
for row in my_list:
    sums.append(row[0] + row[1]) # standard integer "+" operator
sums
```

```python
A = np.array(my_list)
my_sum = A[:,0] + A[:,1] # Vectorial "+" operator
my_sum
```

```python
A.sum(axis=1)
```

### How much faster is Numpy?

```python
# 2D ndarray of shape (10000, 10000) with random float in [0, 1] interval. That's 100M numbers!
A = np.random.rand(10000, 10000)
a = A.tolist()
```

```python
%%time
total = 0
for row in a:
    for n in row:
        total += n
round(total, 2)
```

```python
%%time
round(np.sum(A), 2)
```
