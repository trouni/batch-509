# SQL

```python
import pandas as pd
import sqlite3
conn = sqlite3.connect("../../data/soccer.sqlite")
```

```python
league_df = pd.read_sql(
    '''
    SELECT l.id, l.name, c.name as country_name
    FROM League l
    JOIN Country c ON c.id = l.country_id
    ''', conn)
```
