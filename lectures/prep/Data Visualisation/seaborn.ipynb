```python
tips_df = sns.load_dataset('tips')
```

```python
sns.distplot(tips_df[['total_bill']])
```

```python
sns.countplot(x='time', data=tips_df, hue='sex')
```

```python
sns.catplot(x='time', y='total_bill', data=tips_df)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips_df)
```

```python
sns.regplot(x='total_bill', y='tip', data=tips_df)
```

```python
sns.set(style="darkgrid")
sns.pairplot(tips_df, height=8)
```
