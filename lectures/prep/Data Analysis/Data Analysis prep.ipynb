```python
df = pd.read_csv('countries.csv')
df['Region'] = df['Region'].str.strip()
```

```python
df[(df['Region'].str.contains('AMER')) | (df['Population'] > 100_000_000)]
```

```python
# df['Region'].unique()
df[df['Region'].isin(['EASTERN EUROPE', 'WESTERN EUROPE'])].head(10)
```

```python
df[~df['Region'].isin(['EASTERN EUROPE', 'WESTERN EUROPE'])].head(10)
```

```python
df.Country = df.Country.str.strip()
df.set_index('Country', inplace=True)
```

```python
df.loc['India':'Japan', ['Region']]
```

```python
df.sort_index(ascending=False, inplace=True)
```

```python
df.sort_values(['Region', 'Population'], ascending=[True, False])
```
